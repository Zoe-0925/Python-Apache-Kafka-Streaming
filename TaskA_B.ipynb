Reference Model
We propose the reference model. The climate is the parent collection and the
hotspot is the child collection. The default id of the climate collection is inserted into
the additional field of the hotspot collection.

Example:

'confidence': 86, ‘date’: 27/12/2017, 'surface_temperature_celcius': 67 } ]

Justification:

For data storage and scalability, the reference model yields scalability as the
collections are not likely to reach the memory capacity when the document size
grows fast. Although the document size of both climate and hotspot files are limited,
the document size would increase due to the upcoming streaming data.
For scalability, the reference model handles changes more flexible. When the
climate data and hotspot data are joined by different attributes, in this case, the date
and the location, the modification would be required for only one column, which is
the column in the hotspot collection storing climate id.
For data quality, the reference model yields database atomicity as each collection
holds only a single topic.
For query efficiency, the reference model yields the simplicity of queries by avoiding
the nested dot notation. At data streaming, the producer can append additional
information and create a hierarchical json format without making the queries overcomplicated. Aggregations of a single topic such as group by would also be
easier by only querying one collection.
Although the reference model has the demerit of querying 2 collections when
updating the child collection, the operation can be easily achieved by starting the
database connection once and then inserting or updating multiple records. The cost
is considered lower than the benefits.

```python
#Find climate data on 10th December 2017 .
taskB2a = db.climate_historic.find({"date" : "10/12/2017"})
for result in taskB2a:
    pprint(result)
```

```python
#Find the latitude , longitude, surface temperature (°C) , and confidence when the surface temperature (°C) was between 65 °C and 100 °C .
taskB2b = db.hotspot_historic.find({"surface_temperature_celcius" : {"$gte" : 65, "$lte" : 100}},{"date":0,"datetime":0,"_id":0})
for result in taskB2b:
    pprint(result)
```

```python
#Find date , surface temperature (°C), air temperature (°C), relative humidity and max wind speed on 15th and 16th of December 2017.
taskB2c = db.climate_historic.aggregate([{
"$lookup":
    {
        "from": "hotspot_historic",
        "localField": "date",
        "foreignField": "date",
        "as": "fire"
    }
}
,
{"$unwind":"$fire"}
,
{"$match": {
                "$or": [
                    {"date": "15/12/2017"},                
                    {"date": "16/12/2017"}                
                ]
}}
,
{
    "$project":{"_id":0,"date":1,"fire.surface_temperature_celcius":1,"air_temperature_celcius":1,"relative_humidity":1,"max_wind_speed":1}
}
    
])

for result in taskB2c:
    pprint(result)
```

```python
#Find datetime, air temperature (°C), surface temperature (°C) and confidence when the confidence is between 80 and 100.
taskB2d = db.climate_historic.aggregate([{
"$lookup":
    {
        "from": "hotspot_historic",
        "localField": "date",
        "foreignField": "date",
        "as": "fire"
    }
}
,
    {
        "$match": {"fire.confidence":{"$gte" : 80, "$lte" : 100}}
    }
,
    {
        "$project":{"_id":0,"fire.datetime":1,"air_temperature_celcius":1,"fire.surface_temperature_celcius":1,"fire.confidence":1}
    }
,
    {
        "$unwind":"$fire"
    }
])

for result in taskB2d:
    pprint(result)
```

```python
#Find the top 10 records with the highest surface temperature (°C).
taskB2e = db.climate_historic.aggregate([{
"$lookup":
    {
        "from": "hotspot_historic",
        "localField": "date",
        "foreignField": "date",
        "as": "fire"
    }
    }
    ,
    {
        "$unwind":"$fire"
    }
    ,
    {
       "$sort": {"fire.surface_temperature_celcius" :-1}
    }
    ,
    {
        "$limit":10
    }
])

for result in taskB2e:
    pprint(result)
```

```python
#Find the number of fire in each day. You are required to only display the total number of fire and the date in the output.
taskB2f = db.climate_historic.aggregate([{
"$lookup":
    {
        "from": "hotspot_historic",
        "localField": "date",
        "foreignField": "date",
        "as": "fire"
    }
    }
    ,
    {
        "$unwind":"$fire"
    }
    ,
    {
         "$group": {
             "_id" : {"date" : "$fire.date"},
             "total" : {"$sum" : 1}
         }
    }
])


for result in taskB2f:
    pprint(result)
```

```python
#g. Find the average surface temperature (°C) for each day . You are required to only display average surface temperature (°C) and the date in the output.
taskB2g = db.climate_historic.aggregate([{
"$lookup":
    {
        "from": "hotspot_historic",
        "localField": "date",
        "foreignField": "date",
        "as": "fire"
    }
    }
    ,
    {
        "$unwind":"$fire"
    }
    ,
    {
         "$group": {
             "_id" : {"date" : "$fire.date"},
             "average surface temperature" : {"$avg" : "$fire.surface_temperature_celcius"}
         }
    }
])


for result in taskB2g:
    pprint(result)
```
